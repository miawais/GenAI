**Ollama Embeddings**

```python
#importing ollama embeddings 
from langchain.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma:2b")
)

embeddings
```

```python
# Create a list of documents (strings)
documents = [
    "I am Awais, In this String A is Capital in Awais Name",
    "Hello everyone, I am testing Ollama embeddings"
]

# Call embed_documents with the list of documents
test1 = embeddings.embed_documents(documents)

# Output the embeddings
print(test1)
```
